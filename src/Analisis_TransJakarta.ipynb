### Menangani Missing Value

Kita sudah melihat gambaran kasar anomali di dalam data kita, sekarang, mari kita lihat dan tangani tiap-tiap anomali secara lebih jauh dengan 2 cara berikut: 

1. Mengisi data yang hilang: Pengisian missing value pada setiap kolom dilakukan dengan cara yang berbeda-beda, tergantung pada kesesuaian pengisian nilai yang paling mendekati dan rasional. Rincian pengisian nilai dapat dilihat pada bagian di bawah ini.

2. Menghapus baris sisanya yang memang tidak dapat diisi missing valuenya dengan menggunakan beberapa asumsi dan pertimbangan.

#### `corridorID` dan `corridorName`

Untuk `corridorID` yang tidak ada `corridorName` dan sebaliknya, kita mencoba mengisinya berdasarkan pemetaan yang tersedia.

```python
corridorNameToID = df.dropna(subset=['corridorID', 'corridorName']).drop_duplicates('corridorName').set_index('corridorName')['corridorID'].to_dict()
corridorIDtoName = df.dropna(subset=['corridorID', 'corridorName']).drop_duplicates('corridorID').set_index('corridorID')['corridorName'].to_dict()


df['corridorID'] = df.apply(lambda row: corridorNameToID.get(row['corridorName'], row['corridorID']) if pd.isnull(row['corridorID']) else row['corridorID'], axis=1)

df['corridorName'] = df.apply(lambda row: corridorIDtoName.get(row['corridorID'], row['corridorName']) if pd.isnull(row['corridorName']) else row['corridorName'], axis=1)

df[['corridorID', 'corridorName']].isnull().sum()
```

Setelah melakukan imputasi nilai `corridorID` dan `corridorName`, masih tersisa masing-masing kolom sebanyak 1125 missing value. Saya mengisi missing value yang tersisa dengan memberikan keterangan 'Simulasi' dengan asumsi koridor tersebut adalah simulasi penyesuaian koridor baru yang dilakukan TransJakarta [[1]](https://transjakarta.co.id/transjakarta-lakukan-penyesuaian-terhadap-4-rute/), karena tidak ada kolom lain yang dapat digunakan untuk pemetaan atau mengisi kedua kolom tersebut.

```python
df[['corridorID', 'corridorName']] = df[['corridorID', 'corridorName']].fillna('Simulasi')
```

#### `tapInStops`

Untuk `tapInStops`, kita mencoba mengisinya berdasarkan pemetaan yang tersedia. Untuk imputasi yang keduanya mungkin hilang atau tidak ada pemetaan yang tersedia kita menggunakan imputasi kategori baru dengan mengimplementasikan logika untuk menangani variasi prediksi tapInStops berdasarkan tapInStopsName scikit-learn. Kita akan menggunakan RandomForestClassifier sebagai model prediktif karena kemampuannya yang baik dalam menangani fitur kategorikal dan potensinya untuk model yang kompleks.

```python
# Membuat pemetaan dari tapInStopsName ke tapInStops
mapping = df.dropna(subset=['tapInStops', 'tapInStopsName']).drop_duplicates('tapInStopsName').set_index('tapInStopsName')['tapInStops'].to_dict()

# Menggunakan pemetaan untuk mengisi nilai hilang di tapInStops
df['tapInStops'] = df.apply(lambda row: mapping.get(row['tapInStopsName'], row['tapInStops']) if pd.isnull(row['tapInStops']) else row['tapInStops'], axis=1)
```

```python
# Misal, kita menggunakan LabelEncoder untuk mengubah `tapInStopsName` menjadi numerik
encoder = LabelEncoder()
df['tapInStopsName_encoded'] = encoder.fit_transform(df['tapInStopsName'])

# Memisahkan dataset
dataKnown = df.dropna(subset=['tapInStops'])
dataUnknown = df[df['tapInStops'].isnull()]

# Membagi data_known menjadi set pelatihan dan pengujian
X_train, X_test, y_train, y_test = train_test_split(dataKnown[['tapInStopsName_encoded']], dataKnown['tapInStops'], test_size=0.2, random_state=42)
```

```python
# Inisialisasi dan pelatihan model RandomForestClassifier
model = RandomForestClassifier(n_estimators=100, random_state=42)
model.fit(X_train, y_train)
```

```python
# Evaluasi model
accuracy = model.score(X_test, y_test)
print(f"Akurasi dari imputasi ini sebesar {round(accuracy*100,2)}%")

# Prediksi tapInStops untuk data_unknown
predictedID = model.predict(dataUnknown[['tapInStopsName_encoded']])
print(f'Berikut ID tap in yang akan digunakan untuk imputasi: {predictedID}')

# Mengisi missing value di tapInStops dengan prediksi
df.loc[df['tapInStops'].isnull(), 'tapInStops'] = predictedID
```

**Bagaimana Perubahan dalam Tingkat Kepatuhan Tap Out mempengaruhi Dampak Finansial Pengguna**

Formula:

Pendapatan Tambahan = (Total Transaksi × (1 - Kepatuhan Baru / 100)) × Denda Per Kejadian

```python
totalTransactions = len(df['payAmount'])
currentComplianceRate = 96.45 # Dari analisis sebelumnya, dalam persentase
additionalCostCard = 25000  # Asumsi denda dalam mata uang lokal

# Skenario tingkat kepatuhan
scenarios = {
    "1% Increase": currentComplianceRate + 1,
    "1% Decrease": currentComplianceRate - 1,
    "5% Increase": currentComplianceRate + 5,
    "5% Decrease": currentComplianceRate - 5
}

# Menghitung dampak pendapatan untuk setiap skenario
costImpact = {}
for scenario, currentComplianceRate in scenarios.items():
    missedTapOut = totalTransactions * (1 - (currentComplianceRate / 100))
    additionalCost = missedTapOut * additionalCostCard
    costImpact[scenario] = round(additionalCost)

costImpact
```

Rata-rata Tingkat Kepatuhan Saat Ini: 96.45% dari 37,900 transaksi menghasilkan sekitar 1,344 kejadian tap out yang terlewat.
